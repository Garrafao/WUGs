```python
#each key in clusters dict represents graph and values represents its clusters. need to put it in uncompared clusters and do edge sampling
def uncomp(lemma2graph):
    final_dict = {}
    for lemma,graph in lemma2graph.items():
        samples = []
        uncom =  get_uncompared_clusters(graph,clusters[graph])
        for c1,c2 in uncom:
            cluster1 = clusters[graph][c1]
            cluster2 = clusters[graph][c2]
            uncom_edges = list(product(cluster1,cluster2))
            #print(uncom_edges)
            n = 3 if len(uncom_edges) >= 3 else len(uncom_edges)
            #print(n)
            sample_ind = np.random.choice(len(uncom_edges), size = n, replace = False)
            sample = np.array(uncom_edges)[sample_ind]
            samples = samples + sample.tolist()
        final_dict[lemma] = samples
        #print(len(samples))
                
    return final_dict
```

```python
edge_samp_result = uncomp(graph_dict)
```
