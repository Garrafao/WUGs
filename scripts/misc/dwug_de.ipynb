```python
clusters = {}
for g in graph_dict.values():
    clusters[g],c2n, n2c = get_clusters(g)
```

```python
#each key in clusters dict represents graph and values represents its clusters. need to put it in uncompared clusters and do edge sampling
def uncomp(d):
    final_dict = {}
    for k,v in d.items():
        uncom =  get_uncompared_clusters(v,clusters[v])
        for c1,c2 in uncom:
            cluster1 = clusters[v][c1]
            cluster2 = clusters[v][c2]
            uncom_edges = list(product(cluster1,cluster2))
            #print(uncom_edges)
            n = 5 if len(uncom_edges) >= 5 else len(uncom_edges)
            #print(n)
            sample_ind = np.random.choice(len(uncom_edges), size = n, replace = False)
            sample = np.array(uncom_edges)[sample_ind]
            final_dict[k] = sample
    return final_dict
```

```python
edge_samp_result = uncomp(graph_dict)
```

```python
res_key = list(edge_samp_result.keys())
res_value = list(edge_samp_result.values())
file = open('result.csv', 'w', newline ='')
with file:
    header = ['Identifier1', 'Identifier2', 'lemma']
    writer = csv.DictWriter(file, fieldnames = header)
    writer.writeheader()
    for k,v in edge_samp_result.items():
        for i in v:
            writer.writerow({'Identifier1' : i[0],'Identifier2' : i[1],'lemma' : k})
        
```
